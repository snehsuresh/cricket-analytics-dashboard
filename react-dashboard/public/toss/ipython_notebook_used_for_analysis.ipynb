```python
#loading jsons containing 25 years of data on ODIs, T20s and Test matches from 1991 - 2016
#these jsons were created for my last cricket data story at 
#http://www.wionews.com/cricket/analysis-is-t20-format-ruining-test-cricket-14150

#the jsons are essentially all the scoreboards for all the test, ODI and T20 matches over the past 26 years

#You can see the code I used to generate the jsons at 
#https://nbviewer.jupyter.org/github/shijithpk/2017_04_03_t20_effect/blob/master/t20_notepad_version.ipynb

import json

with open('test_data.json', 'r') as f:
     test_json = json.load(f)

with open('odi_data.json', 'r') as f:
     odi_json = json.load(f)

with open('t20_data.json', 'r') as f:
     t20_json = json.load(f)


```

```python
#just creating a list of years

year_list = range(1991,2017)
```

```python
#combining data from all the three jsons into one CSV
#i kind of visually explore my data in tableau, I'd rather do it that way than with matplotlib or bokeh
#just more familiar with tableau

#note to self: if something doesn't work out, make that cell blank

#thought of limiting the toss stats to matches--be they T20, ODI or Test--between test-playing countries,
#but decided against it, didn't make sense to not use all that data

import json
import csv

headingsx = ['name', 'format','result_text','start_date','team1','team2','toss_text',\
             'toss_winner','toss_winner_choice','venue','venue_country','venue_text','winner','winning_margin','year']

fileg = 'jsons_combined.csv'

with open(fileg, "w") as filem:
    wr = csv.writer(filem, delimiter = ',' , quotechar = '"',lineterminator='\n')
    wr.writerow(headingsx)

for jsonx in [test_json,odi_json,t20_json]:
    for page in jsonx.keys():
        if jsonx[page]['year'] not in ['1990','2017']:
            row_list=[]
            row_list.append(page)
            for head in headingsx[1:]:
                try:
                    jsonx[page][head]
                except:
                    row_list.append('')
                else:
                    row_list.append(jsonx[page][head])
            with open(fileg, "a") as filen:
                wrz = csv.writer(filen, delimiter = ',' , quotechar = '"',lineterminator='\n' )
                wrz.writerow(row_list)

print 'done and done'
```

```python
#so the csv I'd created earlier was a pain to visualise in tableau, so I'm creating a newer csv that is tableau-friendly

#tried to use the pandas library first to create the csv, but gave up after a while.
    #i don't know if i've said this before, but trying to get pandas to do exactly want you want
    #is an exercise in futility, I'd rather just use the plain csv module and manually create my own csvs, than
    #search on stackoverflow for hours in vain for that one code snippet that'll help me with pandas

#In a way, i guess it's the lack of accessible documentation to help beginners like me with not-so-beginner tasks
    #that is putting me off the pandas library

import csv
import pandas as pd

headingsx = ['team', 'opposing_team', 'format', 'year', 'venue_country', 'venue_ground',  \
             'toss_won', 'action_after_toss', 'match_outcome', 'file_name', 'venue_text']

fileg = 'for_tableau.csv'

with open(fileg, "w") as filem:
    wr = csv.writer(filem, delimiter = ',' , quotechar = '"',lineterminator='\n')
    wr.writerow(headingsx)

df2 = pd.read_csv('jsons_combined_v2.csv')

for index,row in df2.iterrows():
    
    #initialising empty lists to which items will get appended
    row_t1 = []
    row_t1_all = []
    row_t2 = []
    row_t2_all = []

    #team_name
    row_t1.append(row['team1'])
    row_t1_all.append('all_nations')
    row_t2.append(row['team2'])
    row_t2_all.append('all_nations')

    #opposing team
    row_t1.append(row['team2'])
    row_t1_all.append(row['team2'])
    row_t2.append(row['team1'])
    row_t2_all.append(row['team1'])
    
    #cricket format
    row_t1.append(row['format'])
    row_t1_all.append(row['format'])
    row_t2.append(row['format'])
    row_t2_all.append(row['format'])
    
    #year
    row_t1.append(row['year'])
    row_t1_all.append(row['year'])
    row_t2.append(row['year'])
    row_t2_all.append(row['year'])
    
    #venue_country
    row_t1.append(row['venue_country'])
    row_t1_all.append(row['venue_country'])
    row_t2.append(row['venue_country'])
    row_t2_all.append(row['venue_country'])
    
    #venue
    row_t1.append(row['venue'])
    row_t1_all.append(row['venue'])
    row_t2.append(row['venue'])
    row_t2_all.append(row['venue'])
    
    #won the toss?
    if row['team1'] == row['toss_winner']:
        row_t1.append('YES')
        row_t1_all.append('YES')
        row_t2.append('NO')
        row_t2_all.append('NO')
    else:
        row_t1.append('NO')
        row_t1_all.append('NO')
        row_t2.append('YES')
        row_t2_all.append('YES')


    #bat_or_field_first, ie. team's first action after toss happened, this covers actions for teams losing toss too
    if row['team1'] == row['toss_winner']:
        row_t1.append(row['toss_winner_choice'])
        row_t1_all.append(row['toss_winner_choice'])
        if row['toss_winner_choice'] == 'bat':  #there's got to be a more elegant way of doing what i just did here
            row_t2.append('field')
            row_t2_all.append('field')
        else:
            row_t2.append('bat')
            row_t2_all.append('bat')
    else:
        if row['toss_winner_choice'] == 'bat':
            row_t1.append('field')
            row_t1_all.append('field')
        else:
            row_t1.append('bat')
            row_t1_all.append('bat')
        row_t2.append(row['toss_winner_choice'])
        row_t2_all.append(row['toss_winner_choice'])
        

    #outcome of match for team - win, loss, other 
    #other means match drawn, tied, cancelled, abandoned etc.
    #if row['winner'] == '':
    if row['winner'] == row['team1']:
        row_t1.append('won')
        row_t1_all.append('won')
        row_t2.append('lost')
        row_t2_all.append('lost')
    elif row['winner'] == row['team2']:
        row_t1.append('lost')
        row_t1_all.append('lost')
        row_t2.append('won')
        row_t2_all.append('won')
    else:
        row_t1.append('other')
        row_t1_all.append('other')
        row_t2.append('other')
        row_t2_all.append('other')

    #file_name where match details are
    row_t1.append(row['name'])
    row_t1_all.append(row['name'])
    row_t2.append(row['name'])
    row_t2_all.append(row['name'])

    #venue_text, this has details on which match is day/night or not
    row_t1.append(row['venue_text'])
    row_t1_all.append(row['venue_text'])
    row_t2.append(row['venue_text'])
    row_t2_all.append(row['venue_text'])

    
    with open(fileg, "a") as filen:
        wrz = csv.writer(filen, delimiter = ',' , quotechar = '"', lineterminator='\n' )
        wrz.writerow(row_t1)
        wrz.writerow(row_t1_all)
        wrz.writerow(row_t2)
        wrz.writerow(row_t2_all)

print 'done and done'
```

```python
#this is the code to make a csv of major test cricket stadiums in New Zealand with latitude and longitude points

#decided to use the OSM nominatim service instead of google's service, seems according to google's policy
    #you can only use it if you plan on displaying the information on google's own maps, which i'm not
    #OSM (open street map), from my understanding of its policy at https://operations.osmfoundation.org/policies/nominatim/ ,
    #is ok with one-time bulk geocoding requests. Think they'll be ok with me geocoding five cricket grounds in New Zealand!


import csv
import geocoder

headingsx = ['number', 'ground', 'lat', 'lon']

fileg = 'd3_03_stadiums_nz.csv'

with open(fileg, "w") as filem:
    wr = csv.writer(filem, delimiter = ',' , quotechar = '"', lineterminator='\n')
    wr.writerow(headingsx)

#placed the stadiums in geographical order, from north to south, will help in labelling on the map later

stad_nz_list = ['Eden Park, Auckland', 'Seddon Park, Hamilton',\
                'Basin Reserve, Wellington', 'Hagley Oval, Christchurch', 'University Oval, Dunedin']

number = 1
for stad in stad_nz_list:
    row_stad = []
    row_stad.append(number)
    row_stad.append(stad)
    coords = geocoder.osm(stad)
    row_stad.append(coords.lat)
    row_stad.append(coords.lng)
    with open(fileg, "a") as filen:
            wrz = csv.writer(filen, delimiter = ',' , quotechar = '"', lineterminator='\n' )
            wrz.writerow(row_stad)
    number += 1

print 'done and done'
```

```python
#part of the code below from http://pandas.pydata.org/pandas-docs/stable/generated/pandas.cut.html and
# https://chrisalbon.com/python/pandas_binning_data.html

bins = [1993, 1998, 2004, 2010, 2016]

group_names = ['1993-1998', '1999-2004', '2005-2010', '2011-2016']

df_39['year_band'] = pd.cut(df_39['year'], bins, labels=group_names, include_lowest=True)

df_39
```

```python
#code below
#from http://stackoverflow.com/questions/34682828/pandas-extracting-specific-selected-columns-from-a-dataframe-to-new-dataframe

df_41 = df_40[['team', 'year_band', 'won_toss_bat_first_percent']].copy()
#code below from https://chrisalbon.com/python/pandas_crosstabs.html
df_41
```

```python
#code below from http://pandas.pydata.org/pandas-docs/stable/reshaping.html

#note to self: crosstabs isn't what you think it is, pandas defines crosstabs as some kind of frequency table
    #I used 'crosstab' as a term for non-flat, excel-like tables. In pandas language you have to use the 'pivot'
    #function to get that


df_42 = df_41.pivot(index='team', columns='year_band', values='won_toss_bat_first_percent')

df_42
```
